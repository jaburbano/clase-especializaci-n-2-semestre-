```python
mensaje = input("Ingresa el mensaje a cifrar: ")

def rot13(texto):
    resultado = ""

    for letra in texto:
        if 'a' <= letra <= 'z':
            resultado += chr((ord(letra) - ord('a') + 13) % 26 + ord('a'))
        elif 'A' <= letra <= 'Z':
            resultado += chr((ord(letra) - ord('A') + 13) % 26 + ord('A'))
        else:
            resultado += letra

    return resultado

mensaje_cifrado = rot13(mensaje)

print("Mensaje original:", mensaje)
print("Mensaje cifrado ROT13:", mensaje_cifrado)
```

```python
import hashlib
md5_hash = hashlib.md5()
mensaje = input("Ingresa el mensaje a cifrar: ")
md5_hash.update(mensaje.encode('utf-8'))
mensaje_cifrado = md5_hash.hexdigest()
print("Mensaje original:", mensaje)
print("Mensaje cifrado MD5:", mensaje_cifrado)
```

```python
import pandas as pd
ruta_archivo = "/content/compras_semana.csv"
df = pd.read_csv(ruta_archivo)
df["total"] = df["precio"] * df["cantidad"]
total_compras = df["total"].sum()
print("Total de compras:", total_compras)
promedio_compras = df["total"].mean()
print("Promedio de compras:", promedio_compras)
mediana_datos = df["total"].median()
print("Mediana de compras:", mediana_datos)
compra_minima = df["total"].min()
stdev_datos = df["total"].std()
print("Desviación estándar de compras:", stdev_datos)
print("Compra mínima:", compra_minima)
var_datos = df["total"].var()
registro_conteo = df["total"].value_counts()
print("Registro conteo:", registro_conteo)
print("Varianza de compras:", var_datos)
print(df.head())
```
